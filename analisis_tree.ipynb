```python
import joblib
from sklearn.tree import export_text
```

```python
with open('model/rfc_3_SMOTE_anxiety_model.pkl', 'rb') as f:
    model = joblib.load(f)
```

```python
first_tree = model.estimators_[0]
```

```python
feature_names = ['avg_velocity', 'max_velocity', 'std_velocity', 'total_distance', 'fixation_ratio', 'unique_area_count']
```

```python
structure = export_text(first_tree, feature_names=feature_names, max_depth=2)
print(structure)
```
